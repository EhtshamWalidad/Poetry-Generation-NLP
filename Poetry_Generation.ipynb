```python
import string

# Define a function to remove punctuation from a string
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
import nltk
from nltk import ngrams
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import ConditionalFreqDist
import random
from string import punctuation
```

```python
from nltk.util import bigrams, trigrams
nltk.download('punkt')
# urdu_punctuations = "؛،؟۔٪«»(){}[]< >!،،-––‘’“”\"'.…،:--—،،"
urdu_punctuations = "؛،؟۔٪«»(){}[]< >!،،-––‘’“”\"'.…،:--—،`'?%"
```

```python

def generateBigramVerse2(poetry,ending_word_list):
    poetry_tokens = word_tokenize(poetry)
    poetry_tokens = [word for word in poetry_tokens if word not in punctuation and word not in urdu_punctuations ]

    # creating ngrams of the words
    unigram_model = ConditionalFreqDist((word,word) for word in poetry_tokens)

    bigram_model = ConditionalFreqDist(bigrams(poetry_tokens))

    rand_word = random.choice(poetry_tokens)
    # rand_word = random.choice(bigram_model.conditions())
    words_length = random.randint(7, 10)

    # getting random word
    verse = [rand_word]

    # prediciting the next word
    for index in range(words_length - 1):

        max_count = 0
        max_words = []

        # if second last word
        if index == words_length - 2:

            # getting highest frequency word
            for key, value in bigram_model[verse[-1]].items():
                if value > max_count:
                   max_words = []
                   max_words.append(key)
                   max_count = value
                elif value == max_count:
                    max_words.append(key)

            flag = False
            for ending_word in ending_word_list:
                for word in max_words:
                    if ending_word[-1] == word[-1]:
                        verse.append(word)
                        flag = True
                        break
                if flag:
                    break

            if not flag:
                verse.append(max_words[0])

            ending_word = verse[-1]

        else:
            # add word to verse
            verse.append(bigram_model[verse[-1]].max())


    return verse, ending_word

stanza_length = 4
verse_length = 4
bigram2_poem = '\n'
ed = ''
ending_word_list = []

po=input()
df = poem.loc[poem['Poet'] == str(po)]
res=df['Poem'].values
res[0]=remove_punctuation(res[0])

for _ in range(stanza_length):
    # ending_word_list = []
    for _ in range(verse_length):
        verse, ed = generateBigramVerse2(res[0],ending_word_list)
        bigram2_poem += " ".join(verse)
        bigram2_poem += '\n'

        if ed not in ending_word_list:
            ending_word_list.append(ed)

    bigram2_poem += '\n'

print(bigram2_poem)
```

```python
import random

stanza_length = 4
verse_length = 8  # Each verse will contain 8 words
verses_per_stanza = 4
trigram_poem = '\n'
ed = ''
ending_word_list = []

po = input()
df = poem.loc[poem['Poet'] == str(po)]
res = df['Poem'].values
res[0]=remove_punctuation(res[0])

def generateBidirectionalTrigramVerse(poem, ending_word_list):
    words = poem.split()
    trigrams = [(words[i], words[i + 1], words[i + 2]) for i in range(len(words) - 2)]

    # Choose a random starting trigram
    current_trigram = random.choice(trigrams)
    verse = [current_trigram[0], current_trigram[1]]

    while len(verse) < verse_length:
        # Find all possible next words based on the current trigram
        next_words = [trigram[2] for trigram in trigrams if trigram[:2] == (current_trigram[1], current_trigram[2])]

        # If there are no possible next words, break the loop
        if not next_words:
            break

        # Choose a random next word
        next_word = random.choice(next_words)
        verse.append(next_word)

        # Update the current trigram
        current_trigram = (current_trigram[1], current_trigram[2], next_word)

    ending_word_list.append(verse[-1])
    return verse, verse[-1]

for _ in range(stanza_length):
    for _ in range(verses_per_stanza):
        verse, ed = generateBidirectionalTrigramVerse(res[0], ending_word_list)
        trigram_poem += " ".join(verse)
        trigram_poem += '\n'

    trigram_poem += '\n'

print(trigram_poem)
```
